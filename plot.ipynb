```python
import matplotlib.pyplot as plt
```

```python
dpca =  [0.09035717682903284, 0.07757722474647674, 0.04388243913666048, 0.028337389688390974, 0.01715885393417379, 0.009156166135083259]
vae = [0.05890114620327949, 0.051679198890924455, 0.04206698015332222, 0.03245740516111255, 0.03151847777888179, 0.031597367078065874]
ppca = [0.11195210995, 0.09659253787, 0.07610484275, 0.0553210123, 0.03552764705, 0.01997905613]
```

```python
#rounding
dpca_rounded = [round(value, 4) for value in dpca]
vae_rounded = [round(value, 4) for value in vae]
ppca_rounded = [round(value, 4) for value in ppca]
print(dpca_rounded, vae_rounded, ppca_rounded)
```

```python
plt.plot([2**(i) for i in range(1, 7)], dpca_rounded, label='PCA', marker='o')
plt.plot([2**(i) for i in range(1, 7)], vae_rounded, label='VAE', marker='x')
plt.plot([2**(i) for i in range(1, 7)], ppca_rounded, label='Probabilistic PCA', marker='s')
plt.xlabel('Latent Dimensions')
plt.ylabel('MSE Loss')
plt.title('MSE Loss Compared')


plt.legend()

plt.show()
```
